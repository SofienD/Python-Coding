**Question 1** 
Write a Python function to find the Max of three numbers.

```python
def max(a,b,c):
    list=[a,b,c]
    list.sort()
    l=len(list)
    max=list[l-1]
    print("le max est:", max)
```

```python
max(25,90,3)
```

**Question 3 :** Write a function that sums the elements of a list of integers
* Write a function that multiplies the elements of an integer list.
* Use the two functions to sum the elements whose position is an even number (0,2,4…) and multiply the rest.
Hint: Consider extracting two lists from a first list 

```python
list=[2,5,7,8,20]

def sum(list):
    l=len(list)
    s=0
    for i in range(0,l,1):
      s=s+list[i]  
    return(s)
```

```python
sum(list)
```

```python
def mult(list):
    l=len(list)
    s=1
    for i in range(0,l,1):
      s=s*list[i]
    return(s)
```

```python
mult(list)
```

```python
def extract(list):
  l=len(list)
  list1=[]
  list2=[]
  for i in range(0,l,1):
    if i%2==0:                     #détermination des indices paires pour la sommation et extraction de la liste list1
      list1.append(list[i])
    else:
        list2.append(list[i])      #détermination des indices impaires pour la multiplication et extraction de la liste list2
  print(sum(list1))
  print(mult(list2))
```

```python
extract(list)
```
